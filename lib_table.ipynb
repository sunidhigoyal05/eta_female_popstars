## Bringing in Artist Work Character Length Table

```python
arwork = pd.read_csv('artist_work_char_len.csv')
arwork 
```

```python
arwork.columns
```

```python
arwork.columns = ['name', 'artist_work', 'character_count']
```

```python
arwork = arwork.set_index('name')
```

```python
arwork
```

```python
LIB = pd.merge(LIB, arwork, on="name")
```

```python
LIB = LIB.drop(["artist_work"], axis=1)
```

```python
LIB = LIB.reset_index()
```

```python
LIB.columns = ['artist',  'nationality', 'genre', 'decade_of_prominence', 'birth_year',
       'instruments', 'character_count']
```

```python
LIB = LIB.set_index('artist')
LIB
```

```python
csv_file = "lib.csv"
LIB.to_csv(csv_file)
```
