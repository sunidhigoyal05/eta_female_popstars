```python
import numpy as np
import pandas as pd
import requests
import json
import os
import dotenv
import sys
sys.tracebacklimit = 0
```

```python
#Created a .env file, and accessing it here 
dotenv.load_dotenv()
```

```python
clientid = os.getenv('geniusclientid')
clientsecret= os.getenv('geniusclientsecret')
accesstoken=os.getenv('geniusaccesstoken')
```

```python
r = requests.get('https://httpbin.org/user-agent')
useragent = json.loads(r.text)['user-agent']
headers = {'User-Agent': useragent,
           'From': 'myp8ma@virginia.edu'}
```

```python
root = 'https://api.genius.com'
endpoint = '/search'
parameters = {'access_token':accesstoken,
              'q':'Adele'}
r = requests.get(root+endpoint,params = parameters, 
                 headers = headers)
r
```

```python
myjson = json.loads(r.text)
myjson['response']['hits'][0]['result']['primary_artist']
```

```python
adele_id=myjson['response']['hits'][0]['result']['primary_artist']['id']
adeleendpoint = myjson['response']['hits'][0]['result']['primary_artist']['api_path']
adeleendpoint
```

```python
endpoint = f'/artists/{adele_id}/songs/'
parameters = {'access_token':accesstoken,
              'q':'Adele',
              'id':adeleendpoint,
              'sort':'popularity',
              'per_page':50,
              'page':'1'
              }
r = requests.get(root+endpoint,params = parameters, 
                 headers = headers)
```

```python
r
```

```python
songsadele=pd.json_normalize(json.loads(r.text)['response']['songs'])
songsadele
```

```python
songsadele = songsadele[songsadele['artist_names'] == 'Adele']
```

```python
songsadele.columns
```

```python
columnstokeep = ['artist_names', 'full_title']
df = songsadele[columnstokeep].copy()
```

```python
songsadele 
```

```python
df
```

```python
df['full_title'] = df['full_title'].str.slice(stop=-9)
df
```

```python
import lyricsgenius 
```

```python
genius = lyricsgenius.Genius(accesstoken, timeout=10000000)
```

```python
for i in range(len(df)):
    # Search for the song using the Genius API
    song = genius.search_song(df.iloc[i, 1], df.iloc[i, 0])
    
    # Check if the song was found
    if song is not None:
        # Print the lyrics without the first three lines
        print("\n".join(song.lyrics.split("\n")[3:-2]))
        
        # Save the title of the song to a text file
        with open("adele_lyrics.txt", "a") as file:
            file.write("'")
            file.write(song.title + "'\n")
            file.write("\n\n")
            file.write("\n".join(song.lyrics.split("\n")[3:-2]))
            file.write("\n\n\n\n")
```
